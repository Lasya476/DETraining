```python
#3. Adding and Modifying Columns 

df['DATE OCC'] = pd.to_datetime(df['DATE OCC'])
```

```python
df['DayOfWeek'] = df['DATE OCC'].dt.day_name()
print(df)
```

```python
df = df.drop(columns=['DayofWeek'])
```

```python
print(df.columns)
```

```python
print(df)
```

```python
crimes_by_day = df.groupby('DayOfWeek').size().reset_index(name='CrimeCount')
print(crimes_by_day)
```

```python
sorted_crimes = crimes_by_day.sort_values(by='CrimeCount', ascending=False)

day_with_most_crimes = sorted_crimes.iloc[0]


print(day_with_most_crimes)
```
